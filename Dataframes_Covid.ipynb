Extracting all three datasets from github, to place them in a dictionary, making it easier to manipulate them, so we end up having a dictionary with 3 dataframes. In this case we're always going to work with updated data.

```
import pandas as pd
import numpy as np

confirmed=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
confirmed.head()
```

```
recovered=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
recovered.head()
```

```
death=pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
death.head()
```

Building a dictionary with 3 datasets.


```
data={'confirmed':confirmed,"deaths":death,"recovered":recovered}
data['confirmed'].head()
```

```
data['deaths'].head()
```

```
data["recovered"].head()
```

The goal now is to remove all useless information, such as state names for instance.

```
for situation, dataset in data.items():
  del data[situation]['Province/State']

```

saving each dataframe in the dictionary of dataframes to a .csv file.


```
import csv

def saver(dictex):
    for key, val in dictex.items():
        val.to_csv("data_{}.csv".format(str(key)))

    with open("keys.txt", "w") as f: #saving keys to file
        f.write(str(list(dictex.keys())))

saver(data)
```
